- Country - страна производства вина.

- Province — провинция или штат, где было произведено вино.
- Region_1 — зона виноделия в провинции или штате (например, долина Напа в Калифорнии).

- Variety— сорт винограда, из которого делают вино (например, Пино Нуар).
- Winery — винодельня, производящая вино.
- Color - цвет вина
- Сountry_to_continent - континент

**Сделаем проверку гипотез**

```python
statistics, p = st.ttest_ind(points_red, points_white, equal_var = False) # stat.ttest_ind метод, который проверяет гипотезу
print('Statistics=%.3f, p=%.3f' % (statistics, p))
# interpret
alpha = 0.05  # 5% 
if p > alpha:
	print('Средние пользовательские рейтинги красного и белого вина одинаковые')
else:
	print('Средние пользовательские рейтинги красного и белого вина разные')
    
```

- В ходе исследования мы получили статистически значимые различия, которые не позволяют нам принять нулевую гипотезу. Следовательно, средние пользовательские рейтинги красного и белого вина разные. 

**Перепроверим гипотезу и проверим ее**

- H0: Средние пользовательские рейтинги красного и белого вина одинаковые.

- H1: Средние пользовательские рейтинги красного вина выше рейтингов белого вина

```python
statistics, p = st.ttest_ind(points_red, points_white, equal_var = False, alternative = 'greater') # stat.ttest_ind метод, который проверяет гипотезу
print('Statistics=%.3f, p=%.3f' % (statistics, p))
# interpret
alpha = 0.05  # 5% 
if p > alpha:
	print('Средние пользовательские рейтинги красного и белого вина одинаковые')
else:
	print('Средние пользовательские рейтинги красного вина выше рейтингов белого вина')
```

- рейтинги красного вина выше рейтингов белого вина
